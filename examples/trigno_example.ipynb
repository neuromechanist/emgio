# Setup and Imports
Import required libraries (os, emgio.EMG, matplotlib.pyplot) and set up the environment for EMG data processing.

```python
# Import required libraries
import os
from emgio import EMG
import matplotlib.pyplot as plt

# Set up the environment for EMG data processing
# (No additional setup required for this example)
```

# Load EMG Data
Load EMG data from a Trigno CSV file using the EMGIO package's from_file method with the 'trigno' importer.

```python
# Load EMG Data

# Sample data path - replace with your actual data path
data_path = 'truncated_trigno_sample.csv'  # Update this path to your Trigno CSV file

if not os.path.exists(data_path):
    print(f"Sample file not found: {data_path}")
    print("Please update the data_path variable with your Trigno CSV file path.")
else:
    # Load the data using Trigno importer
    print("Loading EMG data...")
    emg = EMG.from_file(data_path, importer='trigno')

    # Print available channels
    print("\nAvailable channels:")
    for ch_name, ch_info in emg.channels.items():
        print(f"- {ch_name} ({ch_info['type']})")
        print(f"  Sampling rate: {ch_info['sampling_freq']} Hz")
        print(f"  Unit: {ch_info['unit']}")
```

# Channel Information and Selection
Display available channels and their properties (type, sampling rate, unit). Select and filter EMG channels from the dataset.

```python
# Channel Information and Selection

# Display available channels and their properties (type, sampling rate, unit)
print("\nAvailable channels:")
for ch_name, ch_info in emg.channels.items():
    print(f"- {ch_name} ({ch_info['type']})")
    print(f"  Sampling rate: {ch_info['sampling_freq']} Hz")
    print(f"  Unit: {ch_info['unit']}")

# Select and filter EMG channels from the dataset
emg_channels = [ch for ch, info in emg.channels.items() if info['type'] == 'EMG']
# emg.select_channels(emg_channels)  # TODO: #3 This removes all other channels in place, behavior should change.
```

# Signal Visualization
Create visualization of the EMG signals using matplotlib, showing the first 5 seconds of data with proper titles and labels.

```python
# Signal Visualization

# Plot the first 5 seconds of EMG data
print("\nPlotting EMG signals...")
emg.plot_signals(time_range=(0, 5), channels=emg_channels)
plt.show()
```

# Export to EDF Format
Export the processed EMG data to EDF format and save channel metadata to a TSV file.

```python
# Export to EDF Format

# Define the output path for the EDF file
output_path = 'trigno_emg.edf'

# Export the processed EMG data to EDF format
print(f"\nExporting to EDF: {output_path}")
emg.to_edf(output_path)

# Save channel metadata to a TSV file
channels_metadata_path = f"{os.path.splitext(output_path)[0]}_channels.tsv"
print(f"Channels metadata saved to: {channels_metadata_path}")

print("Export complete!")
```
